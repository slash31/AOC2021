```go
import (
    "fmt"
    "bufio"
    "io/ioutil"
    "strings"
    "os"
    "strconv"
    "sort"
)

type Numeral struct {
    input string
    output string
}

func loadData() []Numeral {
    f, err := os.Open("input_data")
    input := bufio.NewReader(f)
    scanner := bufio.NewScanner(input)
    allLines := []Numeral{}
    for scanner.Scan() {
        components := strings.Split(scanner.Text()," | ")
        element := Numeral{input: components[0], output:components[1]}
        allLines = append(allLines, element)
    }
    return allLines
}

allLines := loadData()
```

```go
// Part 1
total := 0
outLen := map[int]int{}
for i:=0; i < len(allLines); i++ {
    outWords := strings.Split(allLines[i].output," ")
    for j:= 0; j < len(outWords); j++ {
        wl := len(outWords[j])
        if wl == 2 || wl == 3 || wl == 4 || wl == 7 {
            outLen[wl]++
        } 
    }
}

total :=0
for i,v := range outLen {
    total += v
}

fmt.Println(total)
```

```go
// Part 2
type RuneSlice []rune

func (p RuneSlice) Len() int           { return len(p) }
func (p RuneSlice) Less(i, j int) bool { return p[i] < p[j] }
func (p RuneSlice) Swap(i, j int)      { p[i], p[j] = p[j], p[i] }


func union(a,b string) string {
    // all elements of both strings
    tList :=[]rune{}
    tSet := map[int32]int{}
    aR := []rune(a)
    bR := []rune(b) 
    for i := 0; i < len(aR); i++ {
        tSet[aR[i]]=1
    }
    for i := 0; i < len(bR); i++ {
        tSet[bR[i]]=1
    }
    
    for k,_ := range tSet {
        tList = append(tList,int32(k))
    } 
    sort.Sort(RuneSlice(tList))
    return string(tList)
}

func diff(a,b string) string {
    // only elements in a that are not in b
    tList :=[]rune{}
    tSet := map[int32]int{}
    aR := []rune(a)
    bR := []rune(b) 
    for i := 0; i < len(aR); i++ {
        matched := false
        for j := 0; j < len(bR); j++ {
            if aR[i] == bR[j] {
                matched = true
            }
        }
        if matched == false {
            tSet[aR[i]]=1
        }
    }
    for k,_ := range tSet {
        tList = append(tList,int32(k))
    } 
    sort.Sort(RuneSlice(tList))
    return string(tList)
}

func sortString(s string) string {
    r := []rune(s)
    sort.Sort(RuneSlice(r))
    return string(r)
}

func contains (s, t string) bool {
    sA := strings.Split(s,"")
    tA := strings.Split(t,"")
    for i := 0; i < len(tA); i++ {
        match := false
        for j := 0; j < len(sA); j++ {
            if tA[i] == sA[j] {
                match = true
            }
        }
        if !match {
            return false
        }
    }
    return true
}

func mapCode(input,output string) map[string]int {
    codeKeyTemp := map[int]string{}
    codeKey := map[string]int{}
    codes := strings.Split(input," ")
    for i:= 0; i < len(codes); i++ {
        currCode := sortString(codes[i])
        switch len(currCode) {
        case 2:
            codeKeyTemp[1] = sortString(currCode)
        case 3:
            codeKeyTemp[7] = sortString(currCode)
        case 4:
            codeKeyTemp[4] = sortString(currCode)
        case 7:
            codeKeyTemp[8] = sortString(currCode)
        }
    }
    
    for i:= 0; i < len(codes); i++ {
        currCode := sortString(codes[i])
        switch len(currCode) {
        case 5:
            if contains(currCode,diff(codeKeyTemp[4],codeKeyTemp[1])) {
                codeKeyTemp[5] = currCode
            } else if contains(currCode,codeKeyTemp[7]) {
                codeKeyTemp[3] = currCode
            } else {
                codeKeyTemp[2] = currCode
            }
        case 6:
            if contains(currCode,diff(union(codeKeyTemp[4],codeKeyTemp[7]),codeKeyTemp[1])) {
                if contains(currCode,diff(codeKeyTemp[8],codeKeyTemp[1])) {
                    codeKeyTemp[6] = currCode
                } else {
                    codeKeyTemp[9] = currCode
                }
            } else {
                codeKeyTemp[0] = currCode
            }
        }
    }
    
    for k,v := range codeKeyTemp {
        codeKey[v] = k
    }
    
    return codeKey
}
        
total := 0
outLen := map[int]int{}
for i:=0; i < len(allLines); i++ {
    codeKey := mapCode(allLines[i].input, allLines[i].output)
    entry := string("")
    for evIdx,ev := range strings.Split(allLines[i].output," ") {
        inVal := sortString(ev)
        entry += strconv.Itoa(int(codeKey[inVal]))
    }
    total += strconv.Atoi(entry)
}
fmt.Println(total)
```
