```go
// Part 1
type gridTracker struct {
    rows  []int
    cols  []int
}

const size = 5

func generateBoards(rawBoards string) (map[int]*[size][size]int, map[int]*gridTracker) {
    var gameBoards = map[int]*[size][size]int{}
    var markerTracker = map[int]*gridTracker{}
    
    boardNumber := 0
    rowCount := 0
    
    scanner := bufio.NewScanner(strings.NewReader(rawBoards))
    for scanner.Scan() {
        if scanner.Text() == "" {
            boardNumber++
            rowCount = 0
            continue
        }
        if _, a := gameBoards[boardNumber]; !a {
            gameBoards[boardNumber] = &[size][size]int{}
            tmap := gridTracker{
                rows:{size,size,size,size,size},
                cols:{size,size,size,size,size}}
            markerTracker[boardNumber] = &tmap
        }
        rowData := strings.Fields(scanner.Text())
        for i:=0; i < len(rowData); i++ {
            intval, _ := strconv.Atoi(rowData[i])
            gameBoards[boardNumber][rowCount][i] = intval
        }
        rowCount++
    }
    return gameBoards, markerTracker
}

func processDraw(draw int, gameBoards map[int]*[size][size]int, markerTracker map[int]*gridTracker) int {
    for i := 0; i < len(gameBoards); i++ {
        for j := 0; j < len(gameBoards[i]); j++ {
            for k:= 0; k < len(gameBoards[i][j]); k++ {
                if gameBoards[i][j][k] == draw {
                    //fmt.Println("Found",draw,"in board",i,", in pos",j,k)
                    markerTracker[i].rows[j]--
                    if markerTracker[i].rows[j] == 0 {
                        fmt.Println("Winner in board",i,"in row",j,"!")
                        return i
                    }
                    markerTracker[i].cols[k]--
                    if markerTracker[i].cols[k] == 0 {
                        fmt.Println("Winner in board",i,"in col",k,"!")
                        return i
                    }
                }
            }
        }
    }
    return -1
}

func calculateScore(winner *[size][size]int, draws []int) int {
    score := 0
    for i := 0; i < size; i++ {
        for j := 0; j < size; j++ {
            picked := 0
            for _,val := range draws {
                if winner[i][j] == val {
                    picked = 1
                }
            }
            if picked == 0 {
                score += winner[i][j]
            }
        }
    }
    return score * draws[len(draws)-1]
}

gameBoards,markerTracker := generateBoards(boards)
score := 0
for index, draw := range draws {
    winner := processDraw(draw, gameBoards, markerTracker)
    if winner != -1 {
        score = calculateScore(gameBoards[winner], draws[0:index+1])
        break
    }
}
fmt.Println("Total score is",score)
```

```go
// Part 2
type gridTracker struct {
    rows  []int
    cols  []int
}

const size = 5

func generateBoards(rawBoards string) (map[int]*[size][size]int, map[int]*gridTracker) {
    var gameBoards = map[int]*[size][size]int{}
    var markerTracker = map[int]*gridTracker{}
    
    boardNumber := 0
    rowCount := 0
    
    scanner := bufio.NewScanner(strings.NewReader(rawBoards))
    for scanner.Scan() {
        if scanner.Text() == "" {
            boardNumber++
            rowCount = 0
            continue
        }
        if _, a := gameBoards[boardNumber]; !a {
            gameBoards[boardNumber] = &[size][size]int{}
            tmap := gridTracker{
                rows:{size,size,size,size,size},
                cols:{size,size,size,size,size}}
            markerTracker[boardNumber] = &tmap
        }
        rowData := strings.Fields(scanner.Text())
        for i:=0; i < len(rowData); i++ {
            intval, _ := strconv.Atoi(rowData[i])
            gameBoards[boardNumber][rowCount][i] = intval
        }
        rowCount++
    }
    return gameBoards, markerTracker
}

func processDraw(draw int, gameBoards map[int]*[size][size]int, markerTracker map[int]*gridTracker, winnerList map[int]bool) []int {
    allWinners := []int{}
    for i := 0; i < len(gameBoards); i++ {
        if winnerList[i] == true {
            continue
        }
        for j := 0; j < len(gameBoards[i]); j++ {
            for k:= 0; k < len(gameBoards[i][j]); k++ {
                if gameBoards[i][j][k] == draw {
                    //fmt.Println("Found",draw,"in board",i,", in pos",j,k)
                    markerTracker[i].rows[j]--
                    if markerTracker[i].rows[j] == 0 {
                        fmt.Println("Winner in board",i,"in row",j,"!")
                        allWinners = append(allWinners,i)
                    }
                    markerTracker[i].cols[k]--
                    if markerTracker[i].cols[k] == 0 {
                        fmt.Println("Winner in board",i,"in col",k,"!")
                        allWinners = append(allWinners,i)
                    }
                }
            }
        }
    }
    return allWinners
}

func calculateScore(winner *[size][size]int, draws []int) int {
    score := 0
    for i := 0; i < size; i++ {
        for j := 0; j < size; j++ {
            picked := 0
            for _,val := range draws {
                if winner[i][j] == val {
                    picked = 1
                }
            }
            if picked == 0 {
                score += winner[i][j]
            }
        }
    }
    return score * draws[len(draws)-1]
}

gameBoards,markerTracker := generateBoards(boards)
winnerList := make(map[int]bool)
score := 0
for index, draw := range draws {
    winners := processDraw(draw, gameBoards, markerTracker, winnerList)
    for i,w := range winners {
        winnerList[w]=true
        score = calculateScore(gameBoards[w], draws[0:index+1])
        fmt.Println("Total score for board",w,"was",score)
        if len(winnerList) == len(gameBoards) {
            //everyone has won now, break
            break
        }
    }
}
```
