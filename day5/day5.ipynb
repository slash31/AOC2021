```go
import (
    "fmt"
    "bufio"
    "io/ioutil"
    "strings"
    "os"
    "strconv"
)

// Part 1

func strPoint (point []int) string {
    return fmt.Sprintf("%s,%s",strconv.Itoa(point[0]),strconv.Itoa(point[1]))
}

func intPoints (point string) (int,int) {
    x,_ := strconv.Atoi(strings.Split(point,",")[0])
    y,_ := strconv.Atoi(strings.Split(point,",")[1])
    return x,y
}
                                        
func loadData() [][]int {
    f, err := os.Open("input_data")
    data := []string{}
    input := bufio.NewReader(f)
    scanner := bufio.NewScanner(input)
    allLines := [][]int{}
    for scanner.Scan() {
        lineStrings := strings.Split(scanner.Text()," -> ")
        x1,y1 := intPoints(lineStrings[0])
        x2,y2 := intPoints(lineStrings[1])
        lineCoords := []int{x1,y1,x2,y2}
        allLines = append(allLines,lineCoords)
    }
    return allLines
}

func moveTo(a,b int) int {
    if a < b {
        a++
    } else {
        a--
    }
    return a
}

func drawLines (allLines [][]int, pointCounter map[string]int) {
    for i, line := range allLines {
        x1,y1,x2,y2 := line[0],line[1],line[2],line[3]
        pointCounter[strPoint([]int{x1,y1})]++
        if x1 == x2 {
            // this is a vertical line
            for y1 != y2 {
                y1 = moveTo(y1,y2)
                pointCounter[strPoint([]int{x1,y1})]++
            }
        } else if y1 == y2 {
            // this is a horizontal line
            for x1 != x2 {
                x1 = moveTo(x1,x2)
                pointCounter[strPoint([]int{x1,y1})]++
            }
        // condition added for Part 2
        } else {
            // this is a 45-degree diagonal line
            for x1 != x2 {
                x1 = moveTo(x1,x2)
                y1 = moveTo(y1,y2)
                pointCounter[strPoint([]int{x1,y1})]++
            }
        }
    }
}

allLines := loadData()
pointCounter := map[string]int{}
drawLines(allLines, pointCounter)

counter := 0
for i,v := range pointCounter {
    if v >= 2 {
        counter++
    }
}
fmt.Println(counter)
```
