```go
import (
    "fmt"
    "os"
    "bufio"
    "strings"
    "strconv"
)

type Point struct {
    x int
    y int
}

type Instruction struct {
    axis string
    value int
}

func reflect (pointVal int, axisVal int) int {
    if axisVal < pointVal {
        return pointVal - 2*(pointVal-axisVal)
    }
    return pointVal
}

func loadData() (map[Point]bool, []Instruction) {
    f, err := os.Open("test_data")
    input := bufio.NewReader(f)
    scanner := bufio.NewScanner(input)
    points := map[Point]bool{}
    instructions := []Instruction{}
    for scanner.Scan() {
        if scanner.Text() == "" {
            break
        }
        coords := strings.Split(scanner.Text(),",")
        x, _ := strconv.Atoi(coords[0])
        y, _ := strconv.Atoi(coords[1])
        points[Point{x,y}] = true
    }
    for scanner.Scan() {
        foldPosition := strings.Replace(scanner.Text(), "fold along ", "",1)
        entry :=  strings.Split(foldPosition,"=")
        axis := entry[0]
        value,_ := strconv.Atoi(entry[1])
        instructions = append(instructions, Instruction{axis,value})
    }   
    return points, instructions
}

points,instructions := loadData()

func max (a,b int) int {
    if a > b {
        return a
    }
    return b
}

func min (a,b int) int {
    if a < b {
        return a
    }
    return b
}


func pointsToGrid(points map[Point]bool) [][]string {
    maxX := -9999
    maxY := -9999
    for k,_ := range points {
        maxX = max(k.x,maxX)
        maxY = max(k.y,maxY)
    }
    
    var grid = make([][]string, maxY+2)
    for i := range grid {
        grid[i] = make([]string, maxX+2)
    }
    for i := 0; i < maxY+2; i++ {
        for j := 0; j < maxX+2; j++ {
            grid[i][j] = "."
        }
    }
    
    for k,_ := range points {
        grid[k.y][k.x] = "#"
    }
    return grid
}

for i, v := range instructions {
//     if i == 1 {
//         break
//     }
    for p, _ := range points {
        if v.axis == "x" {
            delete(points,p)
            if p.x != v.value {
                points[Point{reflect(p.x,v.value),p.y}]=true
            }
        } else if v.axis == "y" {
            delete(points,p)
            if p.y != v.value {
                points[Point{p.x,reflect(p.y,v.value)}]=true
            }
        }
    }
    
}
pointsGrid := pointsToGrid(points)
for i:=0; i < len(pointsGrid); i++ {
    fmt.Println(pointsGrid[i])
}
```
